```python
import numpy as np
```

```python
from qutip import *
from matplotlib import *
```

```python
H = 2 * np.pi * 0.1 * sigmax()
```

```python
psi0 = basis(2,0)
```

```python
result = mesolve (H, psi0 , np.linspace(0.0,10.0,50),[],[sigmaz(),sigmay()])
# Solve the SE-equantion,Note:H is time indenpendent.
fig , ax = pyplot.subplots()
ax.plot(result.times,result.expect[0]);
ax.plot(result.times,result.expect[1]);
ax.set_xlabel('Time')
ax.set_ylabel('Expectiation Values')
ax.legend(('Sigma-Z','Sigma-y'))# Plot setting
pyplot.show()
```
