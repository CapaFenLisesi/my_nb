```python
%matplotlib inline
import numpy as np
import matplotlib as plt
from qutip import *
```

```python
def dynamic_factor0(t,args ):
    return np.cos(1/2*0.00001*t*t)
def dynamic_factor1(t,args ):
    return np.cos(1/4*0.00001*t*t - np.sqrt(3)/4*args['k2'])
def dynamic_factor2(t,args):
    return np.cos(1/4*0.00001*t*t + np.sqrt(3)/4*args['k2'])

# def moving_k2(w,k1,k2,phi,t):
#     delta1 = 1/4*k1 - np.sqrt(3) / 4 * k2
#     delta2 = 1/4*k1 - np.sqrt(3) / 4 * k2
# #geometry factor
#     exc = basis(3,1)
#     g_1 = basis(3,0)
#     g_2 = basis(3,2)
# #basis of a three-level atom
#     u = 2*t*np.exp(1j*phi)
#     u_dag = 2*t*np.exp(-1j*phi)
# #some factors
#     temp = u*np.cos(1/2*k1)*tensor(exc,g_1)
#     H0 = temp+temp.dag()
# #static hamiltonian
#     temp = u.dag*tensor(exc,g_2)
#     H1 = temp+temp.dag()
#     temp = u*tensor(g_1,g_2)
#     H2 = temp+temp.dag()
#     H = [H0,[H1,dynamic_factor1],[H2,dynamic_factor2]]
    
```

```python
w = 0
k1 = 1
k2 = 1
phi = np.pi/3
t = 1
step = 1000

delta1 = 1/4*k1 - np.sqrt(3) / 4 * k2
delta2 = 1/4*k1 - np.sqrt(3) / 4 * k2
#geometry factor
exc = basis(3,1)
g_1 = basis(3,0)
g_2 = basis(3,2)
#basis of a three-level atom
u = 2*t*np.exp(1j*phi)
u_dag = 2*t*np.exp(-1j*phi)
#some factors
temp = u*np.cos(1/2*k1)*exc*g_1.dag()
H0 = temp+temp.dag()
temp = u_dag*exc*g_2.dag()
H1 = temp+temp.dag()
temp = u*g_1*g_2.dag()
H2 = temp+temp.dag()
H_static=+w*exc*exc.dag()+w*g_1*g_1.dag()+w*g_2*g_2.dag()
#static hamiltonian
H = [H_static,[H0,dynamic_factor0],[H1,dynamic_factor1],[H2,dynamic_factor2]]
#dynamic_hamiltonian
psi0 = (H0 +  H1 * np.cos(1/4*k1 - np.sqrt(3)/4*k2) + H2 * np.cos(1/4*k1 + np.sqrt(3)/4*k2)).eigenstates()[1][2]
#groundstate
F = Qobj([[0,0,2 * t * np.exp(-1j * phi)*np.sin(delta1)*1/4],
            [0,0,-2 * t * np.exp(1j * phi)*np.sin(delta2)*1/4],
            [2 * t * np.exp(1j * phi)*np.sin(delta1)*1/4,-2*t*np.exp(-1j * phi)*np.sin(delta2)*1/4,0]
            ])
#the mearsurement quantity
t_list = np.linspace(0,np.sqrt(k1/(0.001)),step)
result = mesolve(H,psi0,t_list,[],[F],args = {'k2': k2})
```
