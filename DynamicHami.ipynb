```python
import numpy as np
import matplotlib as plt
from qutip import *
```

```python
def my(vx):
    hx_t0 = -99.5
    hz = 1
    H_0 = -hx_t0*sigmax() - sigmaz()*hz
    psi0 = H_0.groundstate()[1]
    H_dynamic = -vx*sigmax()
    H = [H_0,[H_dynamic,'t']]
    step = 1000
    t = np.linspace(0,(0.5-hx_t0)/vx,step)
    output = mesolve(H,psi0,t,[],[sigmay()])
    return output.expect[0][step-1]
```

```python
fig , ax = plt.pyplot.subplots()
```

```python
ax.clear()
```

```python
vx = np.linspace(0.1,5,100)
index = range(0,100)
res = []
for i in index:
    res.append(my(vx[i]))
```

```python
ax.plot(vx,res)
res_1 = (res[1]-res[0])/(vx[1]-vx[0]) * vx
ax.plot(vx,res_1)
ax.set_xlabel('$v_x$')
ax.set_ylabel('$m_s$')
ax.legend(('Time-dependent Simulation','${F}_{yx} v_x$'))# Plot setting
fig
```
